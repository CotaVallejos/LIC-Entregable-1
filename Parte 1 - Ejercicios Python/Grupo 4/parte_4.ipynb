17. **Agregar al Carrito:** Muestra al usuario los productos disponibles y permite que seleccione productos para agregarlos a una lista que simulará ser un "carrito de compras".

```python
catalogo = {
    "manzana": 100,
    "pera": 50,
    "tomate": 40,
    "coca-cola": 1000,
    "vino": 2000,
    "galletas": 500,
}

carrito_de_compras = []

def listar_productos():
    """ Función para listar los productos del carrito de compras """
    for productos in carrito_de_compras:
        for producto, precio in productos.items(): # retorna llaves y valores
            print(f"{producto.title()}: ${precio}")
            
def agregar_producto(producto):
    """ Función para agregar productos al carrito de compras"""
    for prod in catalogo:
        try:
            if producto == prod: # si se encuentra el producto selecionado
                nuevo_producto = {"{prod}": catalogo[prod]}
                carrito_de_compras.append(nuevo_producto) # añadir diccionario con nombre-precio a la lista del carrito
                print("Se ha añadido un producto a su carrito de compras. Su carrito incluye lo siguiente: \n")
                listar_productos()
        except KeyError:
            print("El producto no se encuentra disponible, por favor intente con otro.")
            
comprar = True

while comprar:
    producto = input(f"¿Qué producto desea agregar al carrito? Tenemos disponible lo siguiente: \n"
                    f"Manzana, Pera, Tomate, Coca-Cola, Vino y Galletas. Ingrese el nombre del producto para agregar al carrito y 'q' para terminar la compra.")
    
    if producto.lower() == 'q':
        print("Ha finalizado su compra. Muchas gracias por su preferencia :)")
        comprar = False
    else:
        if producto:
            agregar_producto(producto)
```
